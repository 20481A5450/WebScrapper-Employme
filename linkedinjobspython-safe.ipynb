```python
import time
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException

# Set the LinkedIn job listings URL
url = "https://www.linkedin.com/jobs/data-science-jobs?position=1&pageNum=0"

# Initialize the Chrome driver
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")  # Maximize the browser window
driver = webdriver.Chrome(options=options)

# Navigate to the LinkedIn job listings page
driver.get(url)

# Define a function to click the "See more jobs" button automatically
def click_see_more():
    try:
        see_more_button = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "infinite-scroller__show-more-button"))
        )
        driver.execute_script("arguments[0].scrollIntoView();", see_more_button)
        time.sleep(2)
        see_more_button.click()
    except TimeoutException:
        print("No more 'See more jobs' button found or it's not clickable.")
    except Exception as e:
        print(f"Error clicking 'See more jobs': {e}")

# Define list to store job details
job_data = []

# Scrape the job postings
try:
    while True:  # Keep running until no more "See more jobs" button is found
        job_cards = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, '.base-search-card'))
        )
        for card in job_cards:  # Go through all job postings
            title = card.find_element(By.CSS_SELECTOR, '.base-search-card__title').text
            company = card.find_element(By.CSS_SELECTOR, '.base-search-card__subtitle').text
            location = card.find_element(By.CSS_SELECTOR, '.job-search-card__location').text
            job_link = card.find_element(By.CSS_SELECTOR, '.base-card__full-link').get_attribute('href')

            job_data.append([title, company, location, job_link])

            # Break if 1000 jobs have been scraped
            if len(job_data) >= 1000:
                raise Exception("Reached the limit of 1000 jobs.")

        click_see_more()  # Click "See more jobs" button
        time.sleep(2)  # Sleep to allow the page to load

except Exception as e:
    print(f"An error occurred: {e}")
finally:
    driver.quit()

# Save the scraped data to a CSV file
output_file = 'linkedin_jobs.csv'
with open(output_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Job Title', 'Company', 'Location', 'Job Link'])  # Write the header
    for job in job_data:
        writer.writerow(job)

print(f"{len(job_data)} jobs scraped successfully.")
```
